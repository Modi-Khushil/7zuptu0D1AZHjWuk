```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("/content/ACME-HappinessSurvey2020.csv")

# Splitting the data into features (X) and target (Y)
X = data.drop(columns=["Y"])  # Features
y = data["Y"]  # Target

# Splitting the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.15, random_state=42)

# Train a Random Forest classifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Get feature importances
feature_importances = clf.feature_importances_

# Visualize feature importances
plt.barh(X.columns, feature_importances)
plt.xlabel("Feature Importance")
plt.ylabel("Features")
plt.title("Feature Importance")
plt.show()

# Make predictions on the validation set
y_pred = clf.predict(X_val)

# Evaluate the model
report = classification_report(y_val, y_pred)
print("Classification Report:\n", report)

# Assuming you've selected a few most important features manually from the plot
selected_features = ["X1", "X2", "X5"]

# Train the model with selected features
clf_selected = RandomForestClassifier(random_state=42)
clf_selected.fit(X_train[selected_features], y_train)

# Make predictions on the validation set using the model with selected features
y_pred_selected = clf_selected.predict(X_val[selected_features])

# Evaluate the model with selected features
report_selected = classification_report(y_val, y_pred_selected)
print("Classification Report (Selected Features):\n", report_selected)
```
